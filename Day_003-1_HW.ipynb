### Hints: [隨機產生數值](https://blog.csdn.net/christianashannon/article/details/78867204)

```python
import random
import numpy as np
import pandas as pd

i = np.random.randint(10000000,size=5)
popular = i.tolist()

country = ['A', 'B', 'C', 'D', 'E']

list_labels = ['country', 'popular']

list_cols = [country, popular]

zipped = list(zip(list_labels, list_cols))

world_df = pd.DataFrame(dict(zipped))

print(world_df)

print('max popular is ',country[world_df['popular'].idxmax()] )
```
